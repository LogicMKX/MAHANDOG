# Normal Distribution

```python
# normal distribution curve 
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

# x-axis 
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
         color = "#CA1322", 
        linewidth=2)

plt.title('Normal Distribution of Current Value',
          font = 'Palatino Linotype',
          fontsize = 12)    
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

# Standard Normal Distribution N(0,1)

```python
# Histogram 
plt.hist(df['Z-Score'],
         bins=5,
         density=True,
         alpha=0.8,
         edgecolor='#1C1F24')

mean = df['Z-Score'].mean()
std = df['Z-Score'].std(ddof=1)

# X range based on Z-Score
x = np.linspace(df['Z-Score'].min(),
                df['Z-Score'].max(), 
                100)

p = stats.norm.pdf(x, mean, std)
plt.plot(x, p,
         color = "#CA1322",
          linewidth=2)

plt.title('Normal Distribution of Z-Scores',
          fontname='Palatino Linotype',
          fontsize=12)
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.show()
```
