# sort() and sorted()

```python
numbers = [3,1,5,4,2]
numbers
```

```python
# mutates the list
numbers.sort()
numbers
```

```python
# returns a sorted list
numbers = sorted(numbers, reverse=True)
```

```python
numbers
```

# Slicing

len() - returns the length of the list

```python
len(chars)
```

```python
chars[1:4]
```

```python
chars[0:5:2]
```
