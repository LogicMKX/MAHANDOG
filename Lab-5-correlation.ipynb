```python
# imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Load Dataset

```python
df = pd.read_csv(r'dataset/current-power.csv', 
                delimiter=',')
df
```

```python
# Summary Statistics
df.describe()
```

```python
# Covariance of current and power
df.cov()
```

```python
# Scatter plot 
x = df ['Current']
y = df ['Power']

plt.scatter(x,y)
plt.title('Device Current vs Power Consumption')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
# Correlation Coefficient of Current and Power
df.corr()
```

```python
# Regression Line 
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept
```

```python
# Scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y ,
            color = '#0b56f3',
            alpha = 0.6,
            label = 'Data Point')
```

```python
# Line Plot
plt.plot(x, line,
         color = '#ec4b6c',
         label = f'Regression line(r={r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Air Quality

```python
# Load Dataset
air = pd.read_csv(r'dataset/air-quality.csv', 
                delimiter=',')
air
```

# NOx-Humidity

```python
#Regression Line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (g/m3)')
plt.legend()
plt.show()

# Scatter Plot 
plt.scatter(x,y,
            color = '#0b56f3')
plt.show()
```

```python
# No2 Humidity
x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2 (g/m3)') 
plt.legend()
plt.show()
```

```python
# Scatter Plot 
plt.scatter(x,y,
            color = '#0b56f3')
plt.show()
```

```python
# Temperature vs Humidity
x = air['Temperature']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)


plt.title('Temperature vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Scatter Plot 
plt.scatter(x,y,
            color = '#0b56f3')
plt.show()
```
